```python
def generate_grid_figure(
    gt=None, pred=None, margin=None, only_margin=False, fill_voids=False, figsize=(5, 5)
):
    """
    Generates a grid figure from two 2D arrays and returns the figure object.

    Parameters:
        gt (list of list of int): 2D array of 0s and 1s representing the ground truth grid.
        pred (list of list of int, optional): 2D array of 0s and 1s representing the predicted grid.

    Returns:
        fig: The matplotlib figure object.
    """
    inputs = [gt, pred, margin]
    active_inputs_idxs = [i for i, x in enumerate(inputs) if x is not None]

    if len(active_inputs_idxs) == 0:
        raise ValueError("At least one of gt, pred, or margin must be provided.")

    nrows, ncols = np.array(inputs[active_inputs_idxs[0]]).shape

    if gt is None:
        gt = np.zeros_like(inputs[active_inputs_idxs[0]])
    else:
        gt = np.array(gt)

    if pred is None:
        pred = np.zeros_like(gt)
    else:
        pred = np.array(pred)

    if margin is None:
        margin = np.zeros_like(gt)
    else:
        margin = np.array(margin)

    # Create the figure and axis
    # plt.figure(figsize=(6, 6))
    fig, ax = plt.subplots(figsize=figsize)
    # fig, ax = plt.subplots(figsize=(ncols, nrows))

    # Create the grid with patches for 1s
    # bg_color = np.array((234, 226, 183)) / 255  # light yellow
    bg_color = "white"

    gt_color = np.array((216, 27, 96)) / 255
    pred_color = np.array((30, 136, 229)) / 255

    # margin_color: pastel yellow
    margin_color = np.array((255, 193, 7)) / 255

    true_pos_color = np.array((120, 94, 240)) / 255  #  (gt_color + pred_color) / 2
    # recov_color = (gt_color + margin_color) / 2
    recov_color = np.array((254, 97, 0)) / 255

    # conformal stuff: true positives, recovered_pixels, false positives, false negatives
    true_pos_pixels = np.logical_and(gt == 1, pred == 1)
    # recovered_pixels are all pixels that are in the margin and are also in the ground truth but not in true_pos_pixels
    recovered_pixels = np.logical_and.reduce((gt == 1, pred == 0, margin == 1))
    missed_pixels = np.logical_and.reduce((gt == 1, pred == 0, margin == 0))
    inside_margin = np.logical_and.reduce((pred == 1, margin == 0))

    for i in range(nrows):
        for j in range(ncols):
            alpha = 1.0
            linewidth = 0.5
            edgecolor = "darkgrey"
            color = bg_color
            hatch = None

            if true_pos_pixels[i, j] == 1:
                if len(active_inputs_idxs) == 3:
                    color = bg_color  # do not clutter img: only show what happens with the margin
                    # color = true_pos_color
                    alpha = 0.85
                else:
                    color = true_pos_color
                    alpha = 0.85
            elif recovered_pixels[i, j] == 1:
                color = recov_color
                # add a fine diagonal hatching in black
                hatch = "///"
                # change the color of hatching to black
                edgecolor = "black"
            elif gt[i, j] == 1:
                color = gt_color
                ## to plot GT with black border:
                edgecolor = "black"
                linewidth = 1.5
            elif pred[i, j] == 1:
                if not only_margin:  # and inputs[0] is not None:
                    color = pred_color
                    alpha = 0.75
            elif margin[i, j] == 1:
                color = pred_color
                alpha = 0.5

            if fill_voids:
                if inside_margin[i, j] == 1:
                    color = "gainsboro"
                    # edgecolor = "lightgrey"
                    edgecolor = "gainsboro"
                    alpha = 1.0

            rect = plt.Rectangle(
                (j, nrows - i - 1),
                1,
                1,
                facecolor=color,
                edgecolor=edgecolor,
                linewidth=linewidth,
                alpha=alpha,
                hatch=hatch,
            )
            ax.add_patch(rect)

    for i in range(nrows):
        for j in range(ncols):
            # if np.sum(pred) > 0 and np.sum(margin) > 0 and missed_pixels[i, j] == 1:
            if (np.sum(pred) > 0 and missed_pixels[i, j] == 1) or (
                gt[i, j] == 1 and (pred[i, j] == 0 and margin[i, j] == 0)
            ):
                color = gt_color
                edgecolor = "black"
                # linewidth = 1.5
                linewidth = 2.0
                alpha = 1.0

                rect = plt.Rectangle(
                    (j, nrows - i - 1),
                    1,
                    1,
                    facecolor=color,
                    edgecolor=edgecolor,
                    linewidth=linewidth,
                    alpha=alpha,
                    hatch=hatch,
                )
                ax.add_patch(rect)

    ax.set_xlim(0, ncols + 0.20)
    ax.set_ylim(0 - 0.20, nrows)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.axis("off")

    return fig
```

```python
gt, pred, margin = make_plottanda(niter=2, se_cross=True)

# gt
fig = generate_grid_figure(gt, None, None)
# fig.savefig("grid_01_gt.png")
fig.savefig("grid_01_gt.png", bbox_inches="tight", pad_inches=0.1, dpi=200)
fig.show()

# pred
fig = generate_grid_figure(None, pred, None)
fig.savefig("grid_02_pred.png", bbox_inches="tight", pad_inches=0.1, dpi=200)
fig.show()

# pred vs gt
fig = generate_grid_figure(gt, pred, None)
fig.savefig("grid_03_true_pos.png", bbox_inches="tight", pad_inches=0.1, dpi=200)
fig.show()

# pred + margin
fig = generate_grid_figure(None, pred, margin, False)
fig.savefig("grid_04_margin.png", bbox_inches="tight", pad_inches=0.1, dpi=200)
fig.show()

# La totale
fig = generate_grid_figure(gt, pred, margin, True, True)
fig.savefig(
    "grid_05_recovered_margin.png", bbox_inches="tight", pad_inches=0.1, dpi=200
)
fig.show()
```
